## 1. `concat()` 함수

```python
# 데이터프레임 만들기
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
 
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
```

```python
df1
```

```python
df2
```

```python
# 2개의 데이터프레임을 위 아래 행 방향으로 이어 붙이듯 연결하기 
result1 = pd.concat([df1, df2])
result1
```

```python
# ignore_index=True 옵션 설정하여 기존 index 없애기
result2 = pd.concat([df1, df2], ignore_index=True)
result2
```

```python
# 2개의 데이터프레임을 좌우 열 방향으로 이어 붙이듯 연결하기 
# value가 없으면 NaN(공백)으로 채워짐
result3 = pd.concat([df1, df2], axis=1)
result3
```

```python
# join='inner' 옵션 적용하기 - Inner Join(교집합)
result3_in = pd.concat([df1, df2], axis=1, join='inner')
result3_in
```

```python
# join='inner' 옵션 적용하기 - Outer Join(교집합)는 default
result3_in = pd.concat([df1, df2], axis=1, join='outer')
result3_in
```

## 2.`merge()` 함수

```python
# 주식 데이터를 가져와서 데이터프레임 만들기
df1 = pd.read_excel('./data/stock price.xlsx')
df1
```

```python
df2 = pd.read_excel('./data/stock valuation.xlsx')
df2
```

+ `EPS`(Earning Per Share): 주당순이익
+ `BPS`(Book-value Per Share: 주당순자산가치) = (기업의 현재 순자산) ÷ (총 발행주식수)
+ `PER`(Price Earning Ratio): 주가수익비율 = (주가) ÷ (EPS)
+ `PBR`(Price Book-value Ratio: 주가순자산비율) = (현재주가) ÷ (BPS)


```python
# 데이터프레임 합치기 - 교집합
# default는 Inner Join
merge_inner = pd.merge(df1, df2)
merge_inner
```

```python
print(df1.shape)
print(df2.shape)
```

```python
merge_inner.shape
```

```python
# 데이터프레임 합치기 - 합집합
merge_outer = pd.merge(df1, df2, how='outer')
merge_outer
```

```python
merge_outer.shape
```

```python
# 데이터프레임 합치기 - 합집합
merge_outer_id = pd.merge(df1, df2, how='outer', on='id')
merge_outer_id
```

```python
merge_outer_id.shape
```

```python
merge_left_id = pd.merge(df1, df2, how='left', on='id')
merge_left_id
```

```python
merge_left_id.shape
```

```python
# 데이터프레임 합치기 - 왼쪽 데이터프레임 기준, 키 값 분리
merge_left = pd.merge(df1, df2, how='left', left_on='stock_name', right_on='name')
merge_left
```

```python
merge_left.shape
```

```python
# 데이터프레임 합치기 - 오른쪽 데이터프레임 기준, 키 값 분리
merge_right = pd.merge(df1, df2, how='right', left_on='stock_name', right_on='name')
merge_right
```

```python
merge_right.shape
```

```python
# 불린 인덱싱과 결합하여 원하는 데이터 찾기
price_50000 = df1[df1['price'] < 50000]
price_50000
```

```python
value = pd.merge(price_50000, df2)
value
```

## 3.`join()` 함수

```python
df1 = df1.set_index('id')
df2 = df2.set_index('id')
```

```python
# 데이터프레임 결합(join)
df3 = df1.join(df2)
df3
```

```python
# 데이터프레임 결합(join) - 교집합
df4 = df1.join(df2, how='inner')
df4
```

```python
# 데이터프레임 결합(join) - Left Join
df5 = df1.join(df2, how='left')
df5
```

```python
# 데이터프레임 결합(join) - Right Join
df6 = df1.join(df2, how='right')
df6
```
