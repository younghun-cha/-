```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Lineplot

```python
plt.figure(figsize=([10, 8]))

plt.plot(iris['sepal_length'], iris['sepal_width'], c='g')

plt.title('Sepal Lineplot')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

### 2. Scatter Plot

```python
plt.figure(figsize=([10, 8]))

plt.scatter(iris['sepal_length'], iris['sepal_width'], c='b', marker='x')
plt.scatter(iris['petal_length'], iris['petal_width'], c='r')

plt.title('Iris Scattere plot')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

### 4. Pie Chart

```python
labels = ['IT', 'Marketing', 'Data Science', 'Finance']
values = [500, 156, 300, 510]
explode = [0.01, 0.01, 0.01, 0.1]
colors = ['r', 'b', 'g', 'm'] # colors 인자에 각각 순서대로 color를 지정

plt.pie(values, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%')

plt.title('Classes Rate')

plt.show()

# 이미지 저장
# plt.savefig('saveimage/python01.jpg')
```

### 5. Sub Plot

```python
import math

plt.figure(figsize = ([12, 8]))

x_vals = np.linspace(0, 20, 20)
y_vals = [math.sqrt(i) for i in x_vals]

plt.subplot(2, 2, 1)
plt.plot(x_vals, y_vals, 'bo-')

plt.subplot(2, 2, 2)
plt.plot(x_vals, y_vals, 'rx-')

plt.subplot(2, 2, 3)
plt.plot(x_vals, y_vals, 'g*-')

plt.subplot(2, 2, 4)
plt.plot(x_vals, y_vals, 'go-')

plt.show()

# 이미지 저장
# plt.savefig('python01.jpg')
```
