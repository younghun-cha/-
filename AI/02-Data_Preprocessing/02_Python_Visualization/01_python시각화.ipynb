### 1.2 스타일 지정

```python
x1 = np.linspace(0, 20, 10)
y1 = [math.sqrt(i) for i in x1]

plt.title("Line Chart: Sqrt")

plt.plot(x1, y1, c='r', marker='s', ls='--', mec='g', mew=5, mfc='y')

plt.show()
```

### 1.3 Plot 옵션 지정
* Title
* Labels
* Legends
* Limits
* Grids

```python
x1 = np.linspace(0, 20, 10)
y1 = [math.sqrt(i) for i in x1]

plt.plot(x1, y1, label='Sqaure Root')

plt.title("Plot Options") # title
plt.xlabel('X Values') # x축 라벨
plt.ylabel('Y Values') # y축 라벨
plt.legend(loc=0) # 범례와 위치지정
plt.xlim(5, 20) # x축 범위
plt.ylim(1, 5) # y축 범위
plt.grid(True) # 격자 설정

plt.show()
```

```python
x1 = np.linspace(0, 20, 10)
y1 = [math.sqrt(i) for i in x1]

plt.figure(figsize=([12, 8]))
plt.plot(x1, y1, label='Sqaure Root')

plt.title("Plot Options") # title
plt.xlabel('X Values') # x축 라벨
plt.ylabel('Y Values') # y축 라벨
plt.legend(loc=0) # 범례와 위치지정
plt.xlim(5, 20) # x축 범위
plt.ylim(1, 5) # y축 범위
plt.grid(True) # 격자 설정

plt.show()
```
