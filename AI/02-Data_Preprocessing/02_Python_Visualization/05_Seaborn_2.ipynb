# Python 시각화: Seaborn 2

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 8]

import warnings
warnings.filterwarnings(action='ignore')
```

## 1. Scatter Plot

```python
tips = pd.read_csv('data/tips.csv')
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, color='r')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip',
                data=tips,
                color='r',
                marker='o')
plt.show()
```

## 2. Seaborn 꾸미기

```python
# Grid 넣기 (white, dark, whitegrid, darkgrid, ticks)
#sns.set_style('darkgrid')
sns.set_style('whitegrid')
sns.scatterplot(x='total_bill', y='tip',
                data=tips,
                color='g',
                marker='s')
plt.show()
```

```python
# paper, notebook, talk, poster
sns.set_context('notebook')
sns.scatterplot(x='total_bill', y='tip',
                data=tips,
                color='g',
                marker='s')
plt.show()
```

## 3. Cluster Maps

```python
flights = pd.read_csv('data/flights.csv')
flights.head()
```

```python
pivot = flights.pivot_table(index='month', 
                    columns='year',
                    values='passengers')
```

```python
pivot.head()
```

```python
sns.set_context('notebook')
sns.clustermap(pivot, 
               cmap='coolwarm', 
               linecolor='black', 
               linewidth=0.1)
plt.show()
```

## 4. Pair Grids

```python
plt.rcParams['figure.figsize'] = [20, 12]
sns.pairplot(tips)
plt.show()
```

```python
pgrids = sns.PairGrid(tips)
pgrids.map(plt.scatter)
plt.show()
```

```python
pgrids.map_diag(sns.histplot)
pgrids.map_upper(sns.kdeplot)
pgrids.map_lower(plt.scatter)
plt.show()
```

## 6. Facet Grids

```python
tips.head()
```

```python
fgrid = sns.FacetGrid(data=tips, col='sex', row='time')
fgrid.map(sns.distplot, 'total_bill')
plt.show()
```

```python
fgrid = sns.FacetGrid(data=tips, col='sex', row='time')
fgrid.map(plt.scatter, 'total_bill', 'size')
plt.show()
```
