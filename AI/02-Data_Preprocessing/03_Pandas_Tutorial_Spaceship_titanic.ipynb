## Features 변수명 설명
- PassengerId: 각 승객의 고유 ID(gggg_pp 형식, gggg = 승객이 함께 여행하는 그룹, pp = 그룹 내 번호) 
- HomePlanet: 승객이 출발한 행성
- CryoSleep: 승객이 항해 기간 동안 수면을 일시 중단하도록 선택했는지 여부
- Cabin: 승객이 머물고 있는 객실 번호로 Deck/num/side 형식을 취하며, 여기서 side는 Port의 경우 P 또는 Starboard의 경우 S일 수 있음
- Destination: 승객이 내릴 행성
- Age: 승객의 나이
- VIP: 승객이 항해 중 특별 VIP 서비스를 지불했는지 여부
- RoomService, FoodCourt, ShoppingMall, Spa, VRDeck: 우주선 타이타닉의 다양한 고급 편의 시설 이용 청구금액
- Name: 승객의 성과 이름
- Transported: 승객이 다른 차원으로 운송되었는지 여부로 Target Feature

## 6-2. 결측치 대체: Imputation
- `fillna()` 메서드 사용

```python
# HomePlnanet 결측치는 가장 많은 값을 차지하는 Earth로 대체
titanic['HomePlanet'] = titanic['HomePlanet'].fillna('Earth')
```

```python
# Destination 결측치는 가장 많은 값을 차지하는 TRAPPIST-1e로 대체
titanic['Destination'] = titanic['Destination'].fillna('TRAPPIST-1e')
```

```python
# CryoSleep 결측치는 가장 많은 값을 차지하는 False로 대체
titanic['CryoSleep'] = titanic['CryoSleep'].fillna(False)
```

```python
# VIP 결측치는 가장 많은 값을 차지하는 False로 대체
titanic['VIP'] = titanic['VIP'].fillna(False)
```

```python
# Cabin 결측치는 첫째 값인 B/0/P로 대체
titanic['Cabin'] = titanic['Cabin'].fillna('B/0/P')
```

```python
# 'Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck' 의 평균값을 확인.
# 각각의 결측치에 평균값을 대입.
titanic[['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].mean()
```

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].mean())
```

```python
titanic['RoomService'] = titanic['RoomService'].fillna(titanic['RoomService'].mean())
```

```python
titanic['FoodCourt'] = titanic['FoodCourt'].fillna(titanic['FoodCourt'].mean())
```

```python
titanic['ShoppingMall'] = titanic['ShoppingMall'].fillna(titanic['ShoppingMall'].mean())
```

```python
titanic['Spa'] = titanic['Spa'].fillna(titanic['Spa'].mean())
```

```python
titanic['VRDeck'] = titanic['VRDeck'].fillna(titanic['VRDeck'].mean())
```

```python
# Name의 결측치는 Unknown으로 대체
titanic['Name'] = titanic['Name'].fillna('Unknown')
```

```python
# 결측값 처리 확인
titanic.isna().sum()
```

```python
# 기본 통계치 재확인
titanic.describe()
```

## 7. DataFrame의 컬럼 데이터 엑세스

```python
# Side 컬럼을 "=="으로 추가하기
titanic['Side'] = '=='
titanic.head()
```

```python
# Cabin에서 Side에 입력할 값 확인
titanic['Cabin'].str[-1:]
```

```python
# Cabin 슬라이싱 값을 Side에 입력
titanic['Side'] = np.where(titanic['Cabin'].str[-1:] == 'S', 'S', 'P')
titanic.head()
```

```python
# TotalPayment(총지불금액) 컬럼에 0을 채워 추가하기
titanic['TotalPayment'] = 0
titanic.head()
```

```python
# TotalPayment = RoomService + FoodCourt + ShoppingMall + Spa + VRDeck
titanic['TotalPayment'] = titanic['RoomService'] + titanic['FoodCourt'] + titanic['ShoppingMall'] + titanic['Spa'] + titanic['VRDeck']
titanic.head()
```

```python
# np.where 조건문을 활용하여 Age 값에 따른 AgeGruop 입력
# 1살 이하이면 Baby, 15세 이하이면 Child, 19세 이하히면 Adolescent, 45세 이하이면 Adult, 그 외에는 Elderly
titanic['AgeGroup'] = np.where(
    titanic['Age'] <= 1, 'Baby', np.where(
    titanic['Age'] <= 15, 'Child', np.where(
    titanic['Age'] <= 19, 'Adolescent', np.where(
    titanic['Age'] <= 45, 'Adult', 'Elderly'))))

# 확인
titanic[['Age', 'AgeGroup']]
```

```python
# PassengerId에서 gggg값 대입
titanic['gggg'] = titanic['PassengerId'].str[0:4]
titanic.head()
```

```python
# PassengerId에서 pp값 대입
titanic['pp'] = titanic['PassengerId'].str[5:7]
titanic.head()
```

##  10. apply, lambda로 데이터 가공

```python
# 'Child_Adult' 열을 생성하고 Child를 넣어주고, 만약 15세 이상이면 Adult 문자열 삽입
titanic['Child_Adult'] = titanic['Age'].apply(lambda x: 'Child' if x <= 15 else 'Adult' )
titanic[['Age','Child_Adult']].head(10)
```

```python
# 'Age_Cat' 열을 생성하고 15세 이상은 "adult", 15세 이하는 "Child", 60세 이상은 "Elderly" 채우기
titanic['Age_cat'] = titanic['Age'].apply(lambda x: 'Baby' if x <= 1 else ('Child' if x <= 15 else ('Adolescent' if x <= 19 else ('Adult' if x <= 45 else 'Elderly'))))
titanic.head()
```

```python
# value_counts()로 갯수 확인
titanic['Age_cat'].value_counts()
```

```python
# 이전에 np.where로 만든 AgeGroup과 비교
titanic['AgeGroup'].value_counts()
```
